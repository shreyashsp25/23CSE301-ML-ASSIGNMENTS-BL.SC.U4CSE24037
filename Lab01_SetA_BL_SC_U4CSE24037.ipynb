```python
import numpy as np
import random
import statistics
def count_pairs_with_sum(numbers, target_sum):
    """Counts pairs of elements in a list whose sum equals target_sum."""
    count = 0
    n = len(numbers)
    for i in range(n):
        for j in range(i + 1, n):
            if numbers[i] + numbers[j] == target_sum:
                count += 1
    return count
def calculate_range(numbers):
    """Returns the range (max - min) of a list. Errors if fewer than 3 elements."""
    if len(numbers) < 3:
        return "Range determination not possible"
    return max(numbers) - min(numbers)
def matrix_power(A, m):
    """Returns A raised to the power m."""
    A = np.array(A)
    result = np.identity(A.shape[0], dtype=A.dtype)
    for _ in range(m):
        result = np.matmul(result, A)
    return result
def highest_occurring_char(text):
    """Finds the alphabet character that occurs most frequently in a string."""
    freq = {}
    for ch in text.lower():
        if ch.isalpha():
            freq[ch] = freq.get(ch, 0) + 1
    max_char = max(freq, key=freq.get)
    return max_char, freq[max_char]
def generate_random_list(count, low, high):
    """Generates a list of 'count' random integers between low and high."""
    return [random.randint(low, high) for _ in range(count)]
def compute_stats(numbers):
    """Returns mean, median, and mode of a list of numbers."""
    mean_val = statistics.mean(numbers)
    median_val = statistics.median(numbers)
    mode_val = statistics.mode(numbers)
    return mean_val, median_val, mode_val
pair_count = count_pairs_with_sum([2, 7, 4, 1, 3, 6], 10)
print(f"Q1: Number of pairs with sum 10: {pair_count}")
range_result = calculate_range([5, 3, 8, 1, 0, 4])
print(f"Q2: Range is: {range_result}")
A = [[1, 2], [3, 4]]
m = 3
print(f"Q3: A^{m} ={matrix_power(A, m)}")
char, count = highest_occurring_char("hippopotamus")
print(f"Q4: Maximally occurring character is '{char}' with count {count}")
random_numbers = generate_random_list(25, 1, 10)
mean_val, median_val, mode_val = compute_stats(random_numbers)
print(f"Q5: Generated list: {random_numbers}")
print(f"Q5: Mean: {mean_val}, Median: {median_val}, Mode: {mode_val}")
```
